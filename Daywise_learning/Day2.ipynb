```python
import pandas as pd
df = pd.read_csv("data_cleaning_dataset_1.csv")
df.head() # reading first 5 rows
```

```python
df.duplicated() # check duplicated valu
```

```python
# to remove duplicate value
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()  # checking null values
```

#Handling missing values

```python
# numrical columns---> with meadin


num_cols = ["Age","Marks","Attendance_Percent"]

# With for loop
for col in num_cols:
  df[col]=df[col].fillna(df[col].median())

print(df)
```

```python
# numrical columns---> with meadin


num_cols = ["Age","Marks","Attendance_Percent"]

# Without for loop
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
print(df)
```

```python
#categeorical colums

cat_cols = ["Gender","Department","City","Fee_Paid"]

#with for loop
for cols in cat_cols:
  df[cols] = df[cols].fillna(df[cols].mode([0]))
print(df)
```

```python

df["Email"] = df["Email"].fillna("unknown")
print(df)
```

```python
#Gender Standization

df['Gender'] = df['Gender'].replace({
    "M":"Male",
    "male":"Male",
    "m":"Male",
    "F":"Female",
    "f":"Female",
    "female":"Female"
})
print(df)
```

```python
df['Fee_Paid'] = df['Fee_Paid'].replace({
    "Y":"Yes",
    "y":"Yes",
    "N":"No",
    "n":"No"
})
print(df)
```

# Encoding
* Label Encoding
* Binary Encoding
* One-hot Encoding








```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Course_Level'] = le.fit_transform(df['Course_Level'])
print(df)
```

```python
# Binary Encoding
df['Scholarship'] = df['Scholarship'].map({"Yes":1,"No":0})
df['Fee_Paid'] = df['Fee_Paid'].map({"Yes":1,"No":0})
df['Hostel'] = df['Hostel'].map({"Yes":1,"No":0})
print(df)
```

```python
# One-hot encoding
df = pd.get_dummies(
    df,
    columns = ["Blood_Group"],
    drop_first = True
)
print(df)
```
